# Chi-Squared Test for Categorical Features
# Pearson Coefficient for Numerical Features

```python
from scipy.stats import chi2_contingency, pearsonr

# Function to perform Chi-square test for independence between two categorical features
def chi_square_test(df, feature1, feature2):
    # Create a contingency table
    contingency_table = pd.crosstab(df[feature1], df[feature2])

    # Perform the Chi-square test
    chi2, p, dof, expected = chi2_contingency(contingency_table)

    # Display the results
    print(f"Chi-Square Results for {feature1} and {feature2}\n")
    print(f"Chi-square statistic: {chi2}\n")
    print(f"P-value: {p}\n")
    print(f"Degrees of freedom: {dof}\n")
    print("Expected frequencies table:\n")
    print(expected)

    # Interpretation of the results
    if p < 0.05:
        print("The Chi-square result is statistically significant (reject null hypothesis).")
        print(f"There is evidence of an association between the variables: {feature1} and {feature2}.\n")
    else:
        print("The Chi-square result is not statistically significant (fail to reject null hypothesis).")
        print(f"There is no evidence of an association between the variables: {feature1} and {feature2}.\n")

# Function to calculate Pearson correlation between two numerical features
def pearson_correlation(df, feature1, feature2):
    # Calculate Pearson correlation
    correlation, p_value = pearsonr(df[feature1], df[feature2])

    # Display the results
    print(f"Pearson Correlation Results for {feature1} and {feature2}\n")
    print(f"Correlation coefficient: {correlation}")
    print(f"P-value: {p_value}")

    # Interpretation of the results
    if p_value < 0.05:
        print("The result is statistically significant (reject null hypothesis).")
        print(f"There is a significant correlation between the variables: {feature1} and {feature2}.\n")
    else:
        print("The result is not statistically significant (fail to reject null hypothesis).")
        print(f"There is no significant correlation between the variables: {feature1} and {feature2}.\n")

# Example usage
# Run the Chi-square test for two categorical features
chi_square_test(superstore_df, 'State', 'Postal Code')
chi_square_test(superstore_df, 'Ship Mode', 'Region')
chi_square_test(superstore_df, 'Category', 'Sub-Category')
chi_square_test(superstore_df, 'Region', 'Category')



# Run Pearson correlation for two numerical features
pearson_correlation(superstore_df, 'Sales', 'Profit')
pearson_correlation(superstore_df, 'Quantity', 'Discount')
pearson_correlation(superstore_df, 'Profit', 'Discount')
```

# Approach 1: Chain-of-Thought Prompt

Tried and Tested different ways of Prompting and different parameters. The model needs a simpler prompt with limited instructions.
Thiese prompts generated data, but duplicated the examples.

So, Clear instructions and a simpler way should do generate data.

```python
import re
import pandas as pd
import time
import torch

def generate_cot_prompt(superstore_df):
    example_rows = superstore_df.head(50)  # Limit to 50 examples for clarity
    examples = example_rows.to_string(index=False)
    column_names = superstore_df.columns.tolist()

    base_prompt = f"""
Please generate 10 unique synthetic superstore records that strictly adhere to the column format and value constraints below. The output must be 10 unique CSV rows containing only the column values. Do not include any explanation or intermediate steps.

### Column Format
{', '.join(column_names)}

### Example Records
{examples}

### Constraints for Synthetic Record Generation:
- **Order ID**: Generate a unique Order ID (e.g., "CA-2016-152156"). Allow occasional repetitions for multi-line orders.
- **Order Date**: Generate realistic dates between 2015-01-01 and 2021-12-31, reflecting seasonality.
- **Ship Date**: Must be 1-10 days after the Order Date.
- **Ship Mode**: Randomly choose from 'Standard Class' (60%), 'Second Class' (25%), 'First Class' (10%), 'Same Day' (5%).
- **Customer ID**: Generate unique Customer IDs (e.g., "CG-12520") with occasional repeats to simulate loyal customers.
- **Customer Name**: Ensure alignment with Customer ID.
- **Segment**: Randomly choose from 'Consumer' (50%), 'Corporate' (30%), 'Home Office' (20%).
- **Country**: Always "United States".
- **City, State, Postal Code, Region**: Ensure alignment among these values.
- **Category and Sub-Category**: Generate values that align (e.g., "Furniture", "Chairs").
- **Product Name**: Must align with Sub-Category.
- **Sales, Quantity, Discount, Profit**:
    - Sales: Range $0.44 to $22,638.48.
    - Quantity: Range 1 to 14.
    - Discount: Range 0 to 0.8 (frequent values: 0.0, 0.2).
    - Profit: Range -$6,599.98 to $8,399.98.

### Output Instructions
1. Generate exactly 10 unique rows in CSV format.
2. Do NOT include any explanations, reasoning, or headers.
"""
    return base_prompt


def extract_unique_records(response_text):
    try:
        # Define a pattern for a valid CSV row with exactly 20 fields
        csv_pattern = re.compile(r"^(?:[^,]*,){19}[^,]*$")

        # Split the response into lines
        lines = response_text.strip().split("\n")

        # Collect all valid records
        valid_records = {line.strip() for line in lines if csv_pattern.match(line) and "CA-" in line}

        if valid_records:
            return list(valid_records)  # Return unique rows only
        print("No valid records found in the response.")
        return []
    except Exception as e:
        print(f"Error extracting unique records: {str(e)}")
        return []


def generate_synthetic_data(superstore_df, n_records=10):
    unique_records = set()  # To ensure no duplicates
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)

    attempts = 0
    max_attempts = n_records * 2  # Allow more attempts in case of duplicates

    while len(unique_records) < n_records and attempts < max_attempts:
        try:
            prompt = generate_cot_prompt(superstore_df)
            print(f"\nPrompt for iteration {len(unique_records) + 1}:\n{prompt}")

            inputs = tokenizer(prompt, return_tensors="pt").to(device)
            input_token_len = inputs.input_ids.shape[-1]

            outputs = model.generate(
                **inputs,
                max_new_tokens=1200,
                temperature=1.0,  # Higher temperature for diversity
                top_p=0.95,      # Top-p sampling for varied results
                num_return_sequences=3,  # Generate multiple sequences at once
                pad_token_id=tokenizer.eos_token_id
            )

            generated_texts = [
                tokenizer.decode(output[0][input_token_len:], skip_special_tokens=True)
                for output in outputs
            ]

            for generated_text in generated_texts:
                new_records = extract_unique_records(generated_text)
                for record_text in new_records:
                    if record_text not in unique_records:  # Ensure no duplicates
                        parsed_record = parse_record(record_text)
                        if parsed_record:
                            unique_records.add(record_text)
                            print(f"Added Record: {parsed_record}")
        except Exception as e:
            print(f"Error in generation: {str(e)}")

        attempts += 1

    if not unique_records:
        print("\nNo valid records generated.")
    return list(unique_records)


def parse_record(record_text):
    try:
        fields = record_text.split(",")
        field_names = [
            "Order ID", "Order Date", "Ship Date", "Ship Mode",
            "Customer ID", "Customer Name", "Segment", "Country",
            "City", "State", "Postal Code", "Region", "Product ID",
            "Category", "Sub-Category", "Product Name",
            "Sales", "Quantity", "Discount", "Profit"
        ]

        if len(fields) != len(field_names):
            print(f"Invalid number of fields: {len(fields)} (expected {len(field_names)}): {record_text}")
            return None

        return dict(zip(field_names, [field.strip() for field in fields]))

    except Exception as e:
        print(f"Error parsing record: {str(e)}")
        return None


def records_to_dataframe(records):
    if not records:
        raise ValueError("No valid records generated. Synthetic data generation failed.")

    data = pd.DataFrame(records)

    # Convert numeric columns to appropriate types
    numeric_columns = ["Sales", "Quantity", "Discount", "Profit"]
    for col in numeric_columns:
        data[col] = pd.to_numeric(data[col], errors="coerce")

    # Convert date columns to datetime
    date_columns = ["Order Date", "Ship Date"]
    for col in date_columns:
        data[col] = pd.to_datetime(data[col], errors="coerce")

    return data


if __name__ == "__main__":
    synthetic_records = generate_synthetic_data(superstore_df, n_records=10)
    print(f"Synthetic Records: {synthetic_records}")
    synthetic_df = records_to_dataframe(synthetic_records)
    synthetic_df.to_csv("/content/drive/MyDrive/TCS_GenAI_SDG/Code/synthetic_superstore_data_Llama-3.xlsx", index=False)
    print("Synthetic data saved to 'synthetic_superstore_data_Llama-3.xlsx'")
```

```python
import torch
import pandas as pd
import time


def generate_cot_prompt(superstore_df, current_distributions=None):


    example_rows = superstore_df.head(20)  # Use first 20 rows as examples, which the source data
    examples = example_rows.to_string()
    column_names = superstore_df.columns.tolist()

    base_prompt = f"""
Please generate a synthetic superstore record by thinking through each feature step by step. After your reasoning, provide ONLY the final record on the last line.

Here are the column names and example records from the real dataset:

Column names: {', '.join(column_names)}

Example records:
{examples}
Generate exactly 20 new records that precisely match this format. Each record should have:

Order ID:
Unique Count: 5000+ unique order IDs.
Most Common Order ID occurs fewer than 5 times.
Think: Order IDs should be unique to transactions but may repeat for multi-line orders.
Generate unique IDs with occasional repetitions reflecting realistic transaction scenarios.

Order Date:
Range: From 2015-01-01 to 2021-12-31.
Average Orders Per Day: ~15.
Think: Order dates should reflect seasonality and peak shopping periods (e.g., holidays).
Generate dates with a realistic temporal pattern, ensuring seasonality.

Ship Date:
Range: Order Date + 1 to 10 days.
Shipping Duration: Mean: 4 days, Median: 3 days, Standard Deviation: 1.5 days.
Think: Ship dates should follow order dates closely and reflect different shipping speeds.
Generate shipping durations with a normal distribution and realistic delays.

Ship Mode:
Modes: 'Standard Class' (60%), 'Second Class' (25%), 'First Class' (10%), 'Same Day' (5%).
Think: Ship modes depend on customer urgency and distance to destination.
Generate modes with a weighted distribution reflecting urgency.

Customer ID:
Unique Count: 4000+ unique customers.
Most Common Customer ID occurs fewer than 10 times.
Think: Customer IDs should reflect repeat purchases but mostly unique customers.
Generate IDs with occasional repeats to reflect loyal customers.

Customer Name:
Unique Count: Matches Customer ID count.
Think: Each Customer ID should have a unique corresponding Customer Name.
Generate names with diversity, ensuring alignment with Customer IDs.

Segment:
Categories: 'Consumer' (50%), 'Corporate' (30%), 'Home Office' (20%).
Think: Segment distribution reflects customer types and business demand.
Generate segments with a proportional distribution matching the original data.

Country:
Values: Always "United States."
Think: This dataset focuses solely on US transactions.
Generate "United States" for all rows.

City:
Unique Count: ~500 cities.
Most Common City: Appears ~50 times.
Think: Cities should align with State and Region distribution.
Generate cities with realistic geographic frequency.

State:
Unique Count: 49 states (excluding territories).
Most Common State: Appears ~200 times.
Think: States should align with City and Region values.
Generate states with proportional representation, ensuring alignment with regions.

Postal Code:
Unique Count: Matches city count.
Think: Postal codes should match the cities and states they represent.
Generate postal codes with realistic mapping to City and State.

Region:
Categories: 'West,' 'East,' 'Central,' 'South.'
Think: Regions should align with State and City distributions.
Generate regions with proportional representation across states and cities.

Category:
Categories: 'Furniture,' 'Office Supplies,' 'Technology.'
Think: Categories should reflect demand trends and balance.
Generate categories with a realistic proportional split.

Sub-Category:
Unique Count: ~20 sub-categories.
Think: Sub-categories should align with Category (e.g., 'Chairs' under 'Furniture').
Generate sub-categories ensuring alignment with categories.

Product Name:
Unique Count: 1850 products.
Most Common Product: Appears 48 times.
Think: Products should align with Sub-Category.
Generate products with realistic frequency and alignment to sub-categories.

Sales:
Mean: $229.86, Standard Deviation: $623.25, Range: $0.44 to $22,638.48.
Think: Sales should correlate with Quantity, Discount, and Product Price.
Generate sales values reflecting realistic pricing tiers.

Quantity:
Range: 1 to 14.
Mean: 3.79, Median: 3.
Think: Quantity should align with product type and sales value.
Generate quantities balancing bulk and single-unit purchases.

Discount:
Range: 0 to 0.8 (frequent: 0.0, 0.2).
Mean: 0.156, Median: 0.2.
Think: Discounts should depend on sales strategies and product categories.
Generate discounts with realistic values and frequent promotional rates.

Profit:
Range: -$6,599.98 to $8,399.98.
Mean: $28.66, Standard Deviation: $234.26.
Think: Profit depends on Sales, Quantity, and Discount.
Generate profit values reflecting occasional losses and high profits.

IMPORTANT OUTPUT INSTRUCTIONS:

Requirements for the output:
1. Provide ONLY the CSV line with exactly 20 fields
2. No additional text, explanations, or reasoning
3. No headers
4. Values should be comma-separated
"""
    if current_distributions:
        constraint_prompt = "\nCurrent distributions to consider:\n"
        for feature, dist in current_distributions.items():
            constraint_prompt += f"\n{feature} current distribution:"
            for value, count in dist.items():
                constraint_prompt += f"\n- {value}: {count}"
        base_prompt += constraint_prompt

    return base_prompt

def extract_final_record(response_text):
    try:
        lines = response_text.strip().split("\n")
        for line in reversed(lines):
            if len(line.split(",")) == 20:
                return line.strip()
        return None
    except Exception as e:
        print(f"Error extracting final record: {str(e)}")
        return None

def parse_record(record_text):
    try:
        fields = record_text.split(",")
        field_names = [
            "Order ID", "Order Date", "Ship Date", "Ship Mode",
            "Customer ID", "Customer Name", "Segment", "Country",
            "City", "State", "Postal Code", "Region", "Product ID"
            "Category", "Sub-Category", "Product Name",
            "Sales", "Quantity", "Discount", "Profit"
        ]

        if len(fields) != len(field_names):
            print(f"Invalid number of fields: {len(fields)} (expected {len(field_names)})")
            return None

        return dict(zip(field_names, [field.strip() for field in fields]))

    except Exception as e:
        print(f"Error parsing record: {str(e)}")
        return None

def generate_synthetic_data(superstore_df, n_records):

    records = []
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)

    current_distributions = {
        "ship_mode": {"Standard Class": 0, "Second Class": 0, "First Class": 0, "Same Day": 0},
        "segment": {"Consumer": 0, "Corporate": 0, "Home Office": 0},
        "category": {"Furniture": 0, "Office Supplies": 0, "Technology": 0}
    }

    max_retries = 3
    retry_delay = 2

    for i in range(n_records):
        for retry in range(max_retries):
            try:
                # Generate the prompt
                prompt = generate_cot_prompt(superstore_df, current_distributions)
                print(f"Prompt used:\n{prompt}")

                # Generate output
                inputs = tokenizer(prompt, return_tensors="pt").to(device)
                input_token_len = inputs.shape[-1]
                outputs = model.generate(
                    **inputs,
                    max_new_tokens=1500,
                    temperature=0.7,
                    top_p=0.95,
                    num_return_sequences=1,
                    pad_token_id=tokenizer.eos_token_id
                )
                generated_text = tokenizer.decode(outputs[0][input_token_len:], skip_special_tokens=True)
                print(f"Generated text: {generated_text}")

                # Parse record
                record_text = extract_final_record(generated_text)
                if record_text:
                    print(f"Extracted Record: {record_text}")
                    record = parse_record(record_text)
                    print(f"\n\nParsed Record: {record}")
                    if record:
                        records.append(record)
                        # Update distributions
                        try:
                            current_distributions["ship_mode"][record["Ship Mode"]] += 1
                            current_distributions["segment"][record["Segment"]] += 1
                            current_distributions["category"][record["Category"]] += 1
                        except KeyError:
                            pass
                        break
                else:
                    print(f"Invalid record generated: {generated_text}")

                if retry < max_retries - 1:
                    time.sleep(retry_delay)

            except Exception as e:
                print(f"Error in iteration {i+1}, retry {retry + 1}: {str(e)}")
                if retry < max_retries - 1:
                    time.sleep(retry_delay)
                continue

    if not records:
        print("No valid records generated.")
    return records


def records_to_dataframe(records):
    data = pd.DataFrame(records)

    # Convert numeric columns to appropriate types
    numeric_columns = ["Sales", "Quantity", "Discount", "Profit"]
    for col in numeric_columns:
        data[col] = pd.to_numeric(data[col], errors="coerce")

    # Convert date columns to datetime
    date_columns = ["Order Date", "Ship Date"]
    for col in date_columns:
        data[col] = pd.to_datetime(data[col], errors="coerce")

    return data

# Main execution
if __name__ == "__main__":
    synthetic_records = generate_synthetic_data(superstore_df,n_records=5)
    print(f"synthetic records.", synthetic_records)
    synthetic_df = records_to_dataframe(synthetic_records)
    synthetic_df.to_csv("/content/drive/MyDrive/TCS_GenAI_SDG/Code/synthetic_superstore_data_Llama-3.xlsx", index=False)
    print("Synthetic data saved to 'synthetic_superstore_data_Llama-3.xlsx'")
```
